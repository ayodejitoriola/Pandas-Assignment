```python
import pandas as pd
```

```python
#Aiii. select the following variables and save them in a new dataframe: continent, locations, total_cases, 
# total_deaths, gdp_per_capita, tests_per_case, life_expectancy,female_smokers, male_smokers, diabetes_prevalence
Covid=Covid_subset[["continent", "location", "total_cases", "total_deaths", "gdp_per_capita", "tests_per_case", 
                    "life_expectancy", "female_smokers", "male_smokers", "diabetes_prevalence"]]
Covid
```

```python
#Aiv. Find the missing values in the variables selected
Covid.isna().sum()
```

```python
#B
# The missing values will be dropped due to no value in any of the selected values, 
# we have 205 row and the missing rows are also 205

Covid.drop(['tests_per_case','female_smokers','male_smokers'],axis=1, inplace=True)
```

```python
#C. Find the Continent with the highest total cases and total deaths of coronavirus
Covid.groupby('continent')['total_cases','total_deaths'].max()
```

```python
#D. Which Country has the highest number of deaths?
Covid.groupby('location')['total_deaths'].max()
```

```python
#E. Which Continent has the highest number of deaths and total cases of Covid-19
Covid.groupby('continent').agg({'total_deaths':'max','total_cases':'sum'})
```
